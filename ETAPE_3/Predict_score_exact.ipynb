# Regression Home Team

Le mieux est de faire équipe par équipe, pas les deux d'un coup.
Ici on fait la HomeTeam.
Car le soucis de les mélangers, c'est qu'un algorithme de regression voit pas la nuance entre le score de la HomeTea et de la AwayTeam.

Du coup, pour s'occuper de la HomeTeam, on ne va prendre que sa moyenne de but ainsi que la moyenne de buts encaissés de l'équipe adverse. Mais on ne prend pas le reste comme la moyenne de but adverse (car ça n'influe pas sur le nombre de buts qu'on va marquer).

Bien entendu, pour faire la AwayTeam, on prendra les colonnes inverses (moy_goal_away / moy_goal_conceded_home).

```python
x=data[['puissance_HomeTeam','puissance_AwayTeam','moy_goals_home','moy_goals_conceded_away']]
y=data['goalHomeTeam_FullTime']
data_test= pd.read_csv(filename_test,sep=";",index_col=0)
x_prediction=data_test[['date','status','id_match','homeTeam','awayTeam','idHomeTeam','idAwayTeam',
                        'puissance_HomeTeam','puissance_AwayTeam','moy_goals_home','moy_goals_conceded_away']]
x_prediction
```

```python
x=np.array(x)
y=np.array(y)
x_train,x_valid,y_train,y_valid=train_test_split(x,y,test_size=0.2)

y_train = y_train.astype(np.int)
classes=np.unique(y_train)
nbClasses=len(classes)
print(x_train.shape,y_train.shape)
print(classes)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
## Score du modèle
lr.score(x_valid,y_valid)
```

### Prediction Home Team

```python
unique = x_prediction.values
resultshome = []
i = 0
for i in unique:
    puissance_homeTeam = i[7]
    puissance_awayTeam = i[8]
    moy_goals_home = i[9]
    moy_goals_conceded_away= i[10]
    result =np.round(lr.predict(np.array([[puissance_homeTeam,puissance_awayTeam,moy_goals_home,moy_goals_conceded_away]])))
    resultshome.append(result[0])
```

On nous dit donc que l'équipe "Celta Vigo" a une prédiction de but de 1,15. Ca tombe bien, elle a marqué un seul but dans ce match

Pas besoin de modèle d'apprentissage avec cette méthode mais la justesse des résultats est à vérifier. J'ai fait une dizaine d'exemple et le taux d'erreur m'a l'air plutôt correct. A vérifier
