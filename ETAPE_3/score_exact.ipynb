```python
x=np.array(x)
y=np.array(y)
x_train,x_valid,y_train,y_valid=train_test_split(x,y,test_size=0.2)

y_train = y_train.astype(np.int)
classes=np.unique(y_train)
nbClasses=len(classes)
print(x_train.shape,y_train.shape)
print(nbClasses)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(512, input_dim=6, activation="relu"))
    model.add(Dense(2, input_dim=6, activation="linear"))
    model.add(Dense(2,activation = "softmax"))
    model.compile(loss="mse", optimizer="adam",metrics=['msle'])
    return model
```

```python
model = create_model()
print(model.summary())
```

```python
save = ModelCheckpoint('model_score_exact.h5', monitor = 'val_loss' , verbose = 1 , save_best_only = False)
```

```python
history = model.fit(x_train, y_train,validation_data=(x_valid,y_valid),batch_size=10,epochs=100,callbacks = [save])
```
