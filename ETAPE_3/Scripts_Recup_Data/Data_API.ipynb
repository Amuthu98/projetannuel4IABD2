## Victoires / Défaites consecutives

```python
def recupDate(x):
    annee=x[0:4]
    mois=x[5:7]
    jour=x[8:10]
    date=datetime.datetime(int(annee),int(mois),int(jour))
    return date
```

### Récupération du nom de chaque équipe

```python
def recupTeam(x):
    string=str(x)
    chaine=("'name': ")
    long=len(chaine)
    result=string.find(chaine)
    if result>-1:
        debut=result+long
        i=0
        name=""
        while i==0:
            if string[debut]=='}':
                i=1
                break
            name=name+string[debut]
            debut+=1
    name=name.replace("'","") 
    return name
```

### Récupération du résultat du match

```python
def winner(score):
    string=str(score)
    chaine=("'winner':")
    long=len(chaine)
    result=string.find(chaine)
    if result>-1:
        debut=result+long
        i=0
        winner=""
        while i==0:
            if string[debut]==',':
                i=1
                break
            winner=winner+string[debut]
            debut+=1
    if (winner==" 'DRAW'"):
        winner=0            
    if (winner==" 'HOME_TEAM'"):
        winner=1
    if (winner==" 'AWAY_TEAM'"):
        winner=2
    
    return winner
```

### Comptage des victoires et défaites consécutives

```python
def compte_win(df_verif,name):
    win_cons=0
    for index, row in df_verif.iterrows():
        home=row["homeTeam"]
        away=row["awayTeam"]
        score=row["score"]
        if ((home==name) and (score==1)):
            win_cons+=1
        if ((away==name) and (score==2)):
            win_cons+=1
        if ((home==name) and (score==2)):
            win_cons=0
        if (((away==name) and (score==1)) or (score==0)):
            win_cons=0
    return win_cons



def compte_lose(df_verif,name):
    lose_cons=0
    for index, row in df_verif.iterrows():
        home=row["homeTeam"]
        away=row["awayTeam"]
        score=row["score"]
        if ((home==name) and (score==2)):
            lose_cons+=1
        if ((away==name) and (score==1)):
            lose_cons+=1
        if ((home==name) and (score==1)):
            lose_cons=0
        if (((away==name) and (score==2)) or (score==0)):
            lose_cons=0
    return lose_cons
```

```python
def victoire_consecutives(x,df):
    date=x['date']
    anterieur=df.loc[df['date']< date]    
    home=x["homeTeam"]
    away=x["awayTeam"]
    nb_home=0
    nb_away=0
    if(anterieur.empty): 
        return nb_home,nb_away 
    else:
        verifHome=anterieur.loc[anterieur['homeTeam']==home]
        verifHome2=anterieur.loc[anterieur['awayTeam']==home]
        
        verifAway=anterieur.loc[anterieur['homeTeam']==away]
        verifAway2=anterieur.loc[anterieur['awayTeam']==away]
        
        df_verif=pd.concat([verifHome,verifHome2])
        df_verif=df_verif.sort_values(by = 'date')
        
        df_verif2=pd.concat([verifAway,verifAway2])
        df_verif2=df_verif2.sort_values(by = 'date')
        
        if (df_verif.empty):
                nb_home=-1
        else:
            nb_home=compte_win(df_verif,home)
        
        if(df_verif2.empty):
            nb_away=-1
        
        else:
            nb_away=compte_win(df_verif2,away)
        
    return nb_home,nb_away
```

```python
def defaites_consecutives(x,df):
    date=x['date']
    anterieur=df.loc[df['date']< date]    
    home=x["homeTeam"]
    away=x["awayTeam"]
    nb_home=0
    nb_away=0
    if(anterieur.empty): 
        return nb_home,nb_away 
    else:
        verifHome=anterieur.loc[anterieur['homeTeam']==home]
        verifHome2=anterieur.loc[anterieur['awayTeam']==home]
        
        verifAway=anterieur.loc[anterieur['homeTeam']==away]
        verifAway2=anterieur.loc[anterieur['awayTeam']==away]
        
        df_verif=pd.concat([verifHome,verifHome2])
        df_verif=df_verif.sort_values(by = 'date')
        
        df_verif2=pd.concat([verifAway,verifAway2])
        df_verif2=df_verif2.sort_values(by = 'date')
        
        if (df_verif.empty):
                nb_home=-1
        else:
            nb_home=compte_lose(df_verif,home)
        
        if(df_verif2.empty):
            nb_away=-1
        
        else:
            nb_away=compte_lose(df_verif2,away)
        
    return nb_home,nb_away
```

## Attribution des données au dataframe final

```python
df_final=pd.DataFrame()
df_final['date']=df_finished['utcDate'].apply(lambda x: recupDate(x))
df_final['homeTeam']=df_finished['homeTeam'].apply(lambda x: recupTeam(x))
df_final['awayTeam']=df_finished['awayTeam'].apply(lambda x: recupTeam(x))
df_final['score']=df_finished['score'].apply(lambda x: winner(x))
df_final['win_cons_home'], df_final['win_cons_away']=zip(*df_final.apply(lambda x: victoire_consecutives(x,df_final),axis=1))
df_final['lose_cons_home'], df_final['lose_cons_away']=zip(*df_final.apply(lambda x: defaites_consecutives(x,df_final),axis=1))
df_final.head()
```

```python
def puissance(name,df):
    puiss=df["puissance"].loc[df["name"]==name]
    puiss=puiss.values
    if(puiss):
        return int(puiss)
    else:
        return 100
```
