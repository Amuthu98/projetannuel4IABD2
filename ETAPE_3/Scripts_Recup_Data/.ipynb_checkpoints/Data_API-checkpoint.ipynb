### On numérise les données afin de pouvoir les analyser

### Score


```python
def winner(score):
    string=str(score)
    chaine=("'winner':")
    long=len(chaine)
    result=string.find(chaine)
    if result>-1:
        debut=result+long
        i=0
        winner=""
        while i==0:
            if string[debut]==',':
                i=1
                break
            winner=winner+string[debut]
            debut+=1
    if (winner==" 'DRAW'"):
        winner=0            
    if (winner==" 'HOME_TEAM'"):
        winner=1
    if (winner==" 'AWAY_TEAM'"):
        winner=2
    
    return winner
```

```python
df_final=pd.DataFrame()
df_final['score']=df_finished['score'].apply(lambda x: winner(x))
```

## Puissance-Teams

```python
teams_puiss=pd.read_csv(r"Teams_puiss.csv",sep=";",index_col=0)
teams_puiss
```

```python
def puissance(score,df):
    string=str(score)
    chaine=("'name': ")
    long=len(chaine)
    result=string.find(chaine)
    if result>-1:
        debut=result+long
        i=0
        name=""
        while i==0:
            if string[debut]=='}':
                i=1
                break
            name=name+string[debut]
            debut+=1
    name=name.replace("'","")          
    
    puiss=df["puissance"].loc[df["name"]==name]
    puiss=puiss.values
    if(puiss):
        return int(puiss)
    else:
        return 100
```

```python
df_final['puissance_homeTeam']=df_finished['homeTeam'].apply(lambda x: puissance(x,teams_puiss))
df_final['puissance_awayTeam']=df_finished['awayTeam'].apply(lambda x: puissance(x,teams_puiss))
df_final
```
